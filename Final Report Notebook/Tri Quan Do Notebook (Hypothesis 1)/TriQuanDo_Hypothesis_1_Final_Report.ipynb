3. **Correlation Analysis (Statistical Techniques)**:

  _Model Chosen:_ Linear Regression,  Pearson correlation and Kendall correlation

  > **3.1. Linear Regression:**

  Benefits:

  + It captures linear relationships between variables, i.e. how one variable affects another in a linear manner.
  + It can be used for predicting values of one variable based on the values of another variable.
  + It provides coefficients for each input feature, which can be easily interpreted as the amount of change in the output variable for a unit change in the corresponding input feature.

  Drawbacks:

  + It assumes a linear relationship between variables, so it cannot capture non-linear relationships.
  + It is sensitive to outliers just like Pearson due to the slope calculation.

  > **3.2. Pearson Correlation:**

  Benefits:

  + It captures the linear relationship between two variables.
  + It produces coefficients between -1 and 1, where -1 indicates max negative correlation, 0 indicates no correlation, and 1 indicates max positive correlation. This makes it interpretable and easy to understand the strength and direction of the relationship between two variables.
  + Coefficients are extremely easy to be calculated for a large number of data points, making it scalable for big datasets.

  Drawbacks:
  + It only captures linear relationships between variables and cannot capture nonlinear ones.
  + It calculates means and standard deviations so it is sensitive to outliers.

  > **3.3 Kendall Correlation:**

  Benefits:

  + It can capture monotonic relationships between variables, which means it can capture non-linear relationships that Pearson cannot.
  + It is less sensitive to outliers compared to Pearson, as it is based on the ranks of the data rather than the actual values.
  + It does not make any assumptions about the distribution of data, making it more suitable for non-normal or non-linear data than Pearson.

  Drawbacks:

  + It may not be as sensitive to linear relationships as Pearson. If the relationship between variables is truly linear, Kendall correlation is less efficient and may not provide accurate results.
  + It involves comparing all pairs of data points, which is computationally expensive for large datasets.
  + It does not provide a simple interpretation like Pearson, as it is based on ranks.

```python
def import_crime_data(file_path=file_20_22, expect_col=[], 
                      file_list=[], read_many=False, wb=False):
  """
    This function imports crime data from a CSV file or 
    a list of CSV files. It drops missing values and 
    unnecessary columns from the data and returns 
    a Pandas DataFrame.

    Parameters:
    file_path (str): 
      the path to the CSV file to import
      (default: file_20_22)
    expect_col (list):
      a list of column names to keep in the data
      (default: [])
    file_list (list):
      a list of file paths to import if read_many is True
      (default: [])
    read_many (bool): 
      True if importing multiple files, False if importing a single file
      (default: False)

    Returns:
    A Pandas DataFrame containing the cleaned crime data.
  """
  # When reading multiples file, return a list of frames
  if read_many is True:
    dataFrame_list = []

    for f_name in file_list:
      data_csv = pd.read_csv(f_name)
      dataFrame_list.append(data_csv)

    dataFrames_Final = []
    # Drop unnecessary columns that only retrieve from expected one
    if len(expect_col) > 0:
      for frame in dataFrame_list:
        new_frame = frame.loc[:, expect_col]
        dataFrames_Final.append(new_frame)
    else:
      dataFrames_Final = dataFrame_list

    return dataFrames_Final

  # Case when import 1 single file only
  data_csv = pd.read_csv(file_path)

  if wb:
    data_csv.dropna(inplace=True)                # drop rows missed value
    data_csv.to_csv("wb"+file_path, index=False) # Write back

  # Drop unnecessary columns that only retrieve from expected one
  if len(expect_col) > 0:
    data_csv = data_csv.loc[:, expect_col]

  return data_csv
```

```python
def group_by_col(dataFrame, col_group_name="Year", col_list=[]):
  """
    This function groups the input Pandas DataFrame by 
    a specified column name or a list of column names.
    It returns a Pandas Series object of the counts 
    for each unique value in the groupby column(s).

    Parameters:
    dataFrame (pandas.DataFrame):
      the input Pandas DataFrame to group by
    col_group_name (str):
      the name of the column to group by (default: "")
    col_list (list): 
      a list of column names to group by (default: False)

    Returns:
    A Pandas Series object of the counts for 
    each unique value in the groupby column(s).
  """
  try:
    if len(col_list)==0:
      return dataFrame[col_group_name].value_counts()

    # Performing data count group by for visualization
    my_data = list()
    for column in col_list:
      data_count = dataFrame[column].value_counts()
      my_data.append(data_count)
    return my_data

  except Exception as e:
    print("Internal errors when performing group_by, please check again", str(e))
    return None
```

```python
# Reference: https://www.arduino.cc/reference/en/language/functions/math/map/
def normalize(x, in_min, in_max, out_min, out_max):
  """
    Normalizes a value x from an input range [in_min, in_max] 
    to an output range [out_min, out_max].

    Parameters:
    -----------
    x : float
        The value to be normalized.
    in_min : float
        The minimum value of the input range.
    in_max : float
        The maximum value of the input range.
    out_min: float
        The minimum value of the output range.
    out_max: float
        The maximum value of the output range.

    Returns:
    --------
    float
        The normalized value of x in the output range.
  """
  return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min
```

```python
def statistic_analysis(col="Primary Type", percent=True):
  try:
    # Output most and min criminal type over year 2001 - 2022
    val_counts = criminal_total['Primary Type'].value_counts()
    tot_counts = val_counts.sum()

    # Calculate the percentage of appearance for each unique value
    val_percents = (val_counts/tot_counts)*100

    # Combine the counts and percentages into a single DataFrame
    result_df = pd.concat([val_counts, val_percents], axis=1)
    result_df.columns = ['count', 'percent']

    # Format the percent column to include the percentage symbol
    if percent:
      result_df['percent'] = result_df['percent'].apply(lambda x: '{:.1f}%'.format(x))
      
    return result_df
  except Exception as e:
    print("Internal error when analysis crimnial statistic, please try again")
    return None
```

```python
def label_function(val, min_rat=5.0):
  """
  Function Name: lambda label_function

  Args:
    val: float, value to be formatted as string with percentage
    min_rat: float, minimum percentage value to be considered

  Return:
    Formatted string with percentage, with one decimal point precision,
    only if val >= min_rat. 
    Empty string otherwise.
  """
  if val >= min_rat: return '{:.1f}%'.format(val)
  else: return ''
```

```python
def crime_rate_plot(col="percent", min_rate=5.0, start_ang=90):
  """
  Plots a pie chart of the percent column and adds a legend for slices
  with percentages greater than a specified minimum rate.

  Args:
      col (str): Name of the column contains percentage val.
      min_rate (float) : Minimum percentage rate for which to display a label.
      start_ang (float): Starting angle for the first slice of the pie chart.

  Returns:
      None
  """
  # plot a pie chart of the percent column
  # remove the percentage symbol from the percent column
  result_df['percent_'] = result_df[col].apply(lambda x: float(x.rstrip('%')))

  fig, ax = plt.subplots()
  wedges, labels, autopct = ax.pie(result_df['percent_'], 
                                  labels=None, autopct=label_function,
                                  startangle=start_ang)

  # create a list of legend labels for slices with percentages greater than 5%
  legend_labels = result_df[result_df['percent_'] >= min_rate].index.tolist()

  # create a new legend for the pie chart 
  # that only displays the labels in legend_labels
  ax.legend(wedges, legend_labels, title="Unique Values", 
            loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

  # add a title to the plot
  title_main = "Criminal Status in Chicago over 20 years (2001-2022)\n"
  sub_titles = "Note that the labels only present for percent > 5.0%"
  plt.title(title_main+sub_titles)

  # display the plot
  plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

def randomForestTraining(X, y, test_size=0.2):
  # Split the data into training and testing sets
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=16)

  # Create instance for RandomForestRegressor
  rf_n100 = RandomForestRegressor(n_estimators=100, random_state=16)

  # Fit the Random Forest model
  rf_n100.fit(X_train, y_train)

  # Evaluate the model on the testing data
  r2_score_n100 = rf_n100.score(X_test, y_test)

  # Display R2 score from model training
  print(f"R2 score on set n=100 test_size ({test_size}): {r2_score_n100:.4f}")
```
